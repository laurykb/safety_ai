```python
import importlib
import logging
import pandas as pd
import numpy as np

import train
import evaluate

importlib.reload(train)
importlib.reload(evaluate)

logging.basicConfig(level=logging.INFO)
embedders = ['bert', 'roberta']
models = ['logistic_regression', 'random_forest', 'svm', 'gbm', 'neural_network']
for embedder in embedders :
  print(f"Embedding : {embedder}")
  X_train, X_test, y_train, y_test = train.get_df_split(embedder)
  for model in models :
    print(f"Model : {model}")
    trained_model = train.train_model(model, X_train, y_train)
    y_pred = evaluate.test_model(trained_model, X_test, y_test, "")
    evaluate.balance_accuracy(y_test, y_pred)
    evaluate.mcc(y_test, y_pred)
```
